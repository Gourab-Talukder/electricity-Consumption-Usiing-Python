```python
import matplotlib.pyplot as plt
# Filter the data for Indore City
indore_value = df[df['City'] == 'Indore']
# Create a new DataFrame for the purposes and their respective consumption values
indore_consumption = pd.DataFrame({
    'Purpose': ['Domestic purpose', 'Commercial purpose', 'Public Water Work & Street Light', 'Others'],
    'Consumption (in lakh units)': [
        indore_value['Consumption of Electricity (in lakh units)-Domestic purpose'].sum(),
        indore_value['Consumption of Electricity (in lakh units)-Commercial purpose'].sum(),
        indore_value['Consumption of Electricity (in lakh units)-Public Water Work & Street Light'].sum(),
        indore_value['Consumption of Electricity (in lakh units)-Others'].sum()
    ]
})

# Plotting the consumption of electricity for all purposes using a barplot
plt.figure(figsize=(10, 6))
sns.barplot(data=indore_consumption, x='Purpose', y='Consumption (in lakh units)')
plt.title('Consumption of Electricity in Indore (in lakh units)')
plt.xlabel('Purpose')
plt.ylabel('Consumption (in lakh units)')
plt.show()
```

```python
# Calculate total electricity consumption across all cities
all_cities = df['Consumption of Electricity (in lakh units)-Total Consumption'].sum()

print(f"Total electricity consumption by Indian cities: {all_cities} lakh units")
```

```python
# Finding the city with maximum electricity consumption for Industrial Purpose
maxmimum = df[df['Consumption of Electricity (in lakh units)-Industry purpose'] == df['Consumption of Electricity (in lakh units)-Industry purpose'].max()]['City'].iloc[0]

# Finding the city with minimum electricity consumption for Industrial Purpose
minimum = df[df['Consumption of Electricity (in lakh units)-Industry purpose'] == df['Consumption of Electricity (in lakh units)-Industry purpose'].min()]['City'].iloc[0]

print(f"City with maximum electricity consumption for Industrial purpose: {maxmimum}")
print(f"City with minimum electricity consumption for Industrial purpose: {minimum}")
```

```python
# Group by year and sum the total electricity consumption
ax = sns.countplot(data = df, x = 'Year')
sns.set(rc={'figure.figsize':(15, 6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Trend Analyze 
df = pd.read_csv("electricity-consumption-india-city-level-2017-19.csv", encoding='unicode_escape')
df.dropna(inplace=True)
df['result'] = df['Consumption of Electricity (in lakh units)-Domestic purpose'] - df['Consumption of Electricity (in lakh units)-Industry purpose'].astype('int')
df.head(5)
df.sort_values(by='result', ascending=False)
```

```python
# Plot the trend over the years
sns.set(rc={'figure.figsize':(15, 6)})
sns.lineplot(data=df, x='Year', y='result')
```
